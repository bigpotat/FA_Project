```python
def return_seriers_df(data):
  pct_change=data["Adj Close"].pct_change()
  ret_series_adjclose=((1+pct_change).cumprod())-1
  return ret_series_adjclose
```

```python
def reset_index(data):
  return data.reset_index(drop = True)
```

```python
fsly_return=reset_index(return_seriers_df(fsly_df))
immr_return=reset_index(return_seriers_df(immr_df))
mttr_return=reset_index(return_seriers_df(mttr_df))
fb_return=reset_index(return_seriers_df(fb_df))
rblx_return=reset_index(return_seriers_df(rblx_df))
meta_return=reset_index(return_seriers_df(meta_df))
u_return=reset_index(return_seriers_df(u_df))
```
